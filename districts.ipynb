```python
import pandas as pd
import requests as req
import json
import os
import csv

response = req.get('https://api.covid19india.org/v2/state_district_wise.json')

os.system('clear')

json1 = json.loads(response.text)

rows = []

for row in json1:
  districtData = row['districtData']
  
  for data in districtData:
    data['state'] = row['state']
    rows.append(data)

DF = pd.DataFrame(rows)

# print(json[1]['districtData'])


DF1=DF.drop("lastupdatedtime",axis=1)
DF2=DF1.drop("delta",axis=1)
data_file = open('data_file.csv', 'w')
csv_writer = csv.writer(data_file) 
DF2.to_csv('Districts.csv', index=False)
```
